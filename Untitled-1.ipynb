```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Dense, Flatten, Input
```

```python

mat_data = scipy.io.loadmat('Xtrain.mat')

key = [k for k in mat_data.keys() if not k.startswith('__')][0]
data = mat_data[key].flatten()

#scaling to 0-1 range
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data.reshape(-1, 1)).flatten()
print(data_scaled.shape)
```

```python
def create_dataset(dataset, look_back=10):
    X, Y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:(i + look_back)])
        Y.append(dataset[i + look_back])
    return np.array(X), np.array(Y)

# change this for optimizing
look_back = 64

X, y = create_dataset(data_scaled, look_back)
X = X.reshape((X.shape[0], X.shape[1], 1))
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

```python
# 80/20 split
train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print(f"Training samples: {len(X_train)}, Testing samples: {len(X_test)}")
```

```python
# TCN model
model = Sequential([
    Input(shape=(look_back, 1)),
    # finetune filter amount
    Conv1D(filters=32, kernel_size=3, padding='causal', dilation_rate=1, activation='relu'),
    Conv1D(filters=32, kernel_size=3, padding='causal', dilation_rate=2, activation='relu'),
    Conv1D(filters=32, kernel_size=3, padding='causal', dilation_rate=4, activation='relu'),
    Conv1D(filters=32, kernel_size=3, padding='causal', dilation_rate=8, activation='relu'),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(
    X_train, y_train, 
    epochs=50, 
    batch_size=32, 
    validation_data=(X_test, y_test),
    verbose=1
)

plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.legend()
plt.title('mdel loss during training')
plt.show()
```

```python

y_pred_scaled = model.predict(X_test)

y_pred = scaler.inverse_transform(y_pred_scaled) #scale back to original values
y_test_real = scaler.inverse_transform(y_test.reshape(-1, 1))


mae = mean_absolute_error(y_test_real, y_pred)
print(f"Local Test Set MAE: {mae:.4f}")
mse = mean_squared_error(y_test_real, y_pred)
print(f"Local Test Set MSE: {mse:.4f}")


plt.figure(figsize=(12, 5))
plt.plot(y_test_real, label='Real Measurements', color='blue', alpha=0.6)
plt.plot(y_pred, label='TCN Predictions', color='red', linestyle='dashed')
plt.title('TCN: predicted vs real values (test set)')
plt.legend()
plt.show()
```

```python
current_window = data_scaled[-look_back:].reshape(1, look_back, 1)

recursive_predictions_scaled = []

for _ in range(200):
    next_pred = model.predict(current_window, verbose=0)[0][0]
    recursive_predictions_scaled.append(next_pred)
    current_window = np.append(current_window[:, 1:, :], [[[next_pred]]], axis=1)


recursive_predictions = scaler.inverse_transform(np.array(recursive_predictions_scaled).reshape(-1, 1))

plt.figure(figsize=(14, 5))
plt.plot(np.arange(len(data)), data, label='Historical Data', color='blue')
plt.plot(np.arange(len(data), len(data) + 200), recursive_predictions, label='200-Step Recursive Forecast', color='green')
plt.title('Recursive Forecasting of Next 200 Points')
plt.legend()
plt.show()
```

```python
import itertools
from tensorflow.keras.optimizers import Adam

#test different combinations of hyperparameters 
param_grid = {
    'look_back': [32, 64, 128],             
    'filters': [8, 16, 32, 64],               
    'learning_rate': [0.002, 0.001, 0.0075, 0.0005]   
}


keys, values = zip(*param_grid.items())
combinations = [dict(zip(keys, v)) for v in itertools.product(*values)]

best_mae = float('inf')
best_params = None

for idx, params in enumerate(combinations):
    print(f"\n--- Model {idx+1}/{len(combinations)} ---")
    print(f"Trainen met parameters: {params}")

    X_curr, y_curr = create_dataset(data_scaled, params['look_back'])
    X_curr = X_curr.reshape((X_curr.shape[0], X_curr.shape[1], 1))

    train_size_curr = int(len(X_curr) * 0.8)
    X_train_c, X_test_c = X_curr[:train_size_curr], X_curr[train_size_curr:]
    y_train_c, y_test_c = y_curr[:train_size_curr], y_curr[train_size_curr:]

    model = Sequential([
        Input(shape=(params['look_back'], 1)),
        Conv1D(filters=params['filters'], kernel_size=3, padding='causal', dilation_rate=1, activation='relu'),
        Conv1D(filters=params['filters'], kernel_size=3, padding='causal', dilation_rate=2, activation='relu'),
        Conv1D(filters=params['filters'], kernel_size=3, padding='causal', dilation_rate=4, activation='relu'),
        Conv1D(filters=params['filters'], kernel_size=3, padding='causal', dilation_rate=8, activation='relu'),
        Flatten(),
        Dense(params['filters'], activation='relu'),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=params['learning_rate']), loss='mse')

    model.fit(X_train_c, y_train_c, 
              epochs=100, 
              batch_size=32,
              validation_data=(X_test_c, y_test_c), 
              verbose=0) 

    y_pred_scaled = model.predict(X_test_c, verbose=0)
    y_pred = scaler.inverse_transform(y_pred_scaled)
    y_test_real = scaler.inverse_transform(y_test_c.reshape(-1, 1))

    mae = mean_absolute_error(y_test_real, y_pred)
    print(f"Resultaat MAE: {mae:.4f}")

    if mae < best_mae:
        best_mae = mae
        best_params = params
        model.save('beste_automatische_model.keras')

print(f"Beste MAE was {best_mae:.4f}")
print(f"Met parameters: {best_params}")
```

```python
from tensorflow.keras.models import load_model

model_to_use = load_model('beste_automatische_model_64lb_32f_0.001l.keras')

def backtest_recursive_model(start_index, model, look_back, steps_to_predict=200):

    print(f"\n backtest vanaf datapunt {start_index} ")
    current_window = data_scaled[start_index - look_back : start_index].reshape(1, look_back, 1)
    
    recursive_preds_scaled = []
    for _ in range(steps_to_predict):
        next_pred = model.predict(current_window, verbose=0)[0][0]
        recursive_preds_scaled.append(next_pred)
        current_window = np.append(current_window[:, 1:, :], [[[next_pred]]], axis=1)
        

    preds_real = scaler.inverse_transform(np.array(recursive_preds_scaled).reshape(-1, 1)).flatten()
    actual_data = data[start_index : start_index + steps_to_predict]
    history_plot = data[start_index - 100 : start_index] 
    
    
    plt.figure(figsize=(12, 5))
    
    plt.plot(range(start_index - 100, start_index), history_plot, color='gray', label='Historische Context')
    plt.plot(range(start_index, start_index + steps_to_predict), actual_data, color='blue', linewidth=2, label='Echte Data (Doel)')
    plt.plot(range(start_index, start_index + steps_to_predict), preds_real, color='red', linestyle='dashed', linewidth=2, label='TCN Recursieve Voorspelling')
    plt.axvline(x=start_index, color='black', linestyle='dotted', label='Startpunt Recursie')
    
    
    temp_mae = mean_absolute_error(actual_data, preds_real)
    
    plt.title(f'Backtest 200 stappen vanaf index {start_index} (MAE: {temp_mae:.2f})')
    plt.legend()
    plt.show()





backtest_recursive_model(start_index=300, model=model_to_use, look_back=64)
backtest_recursive_model(start_index=550, model=model_to_use, look_back=64)
backtest_recursive_model(start_index=650, model=model_to_use, look_back=64)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
test_mat = scipy.io.loadmat('Xtest.mat')

test_key = [k for k in test_mat.keys() if not k.startswith('__')][0]
test_data = test_mat[test_key].flatten()


actual_test_values = test_data[:200] 
model = load_model('beste_automatische_model_64lb_32f_0.001l.keras')
look_back = 64 
current_window = data_scaled[-look_back:].reshape(1, look_back, 1)


recursive_predictions_scaled = []

print("Generating 200 recursive predictions...")
for _ in range(200):
    next_pred = model.predict(current_window, verbose=0)[0][0]
    recursive_predictions_scaled.append(next_pred)
    current_window = np.append(current_window[:, 1:, :], [[[next_pred]]], axis=1)


predicted_values = scaler.inverse_transform(np.array(recursive_predictions_scaled).reshape(-1, 1)).flatten()
mae = mean_absolute_error(actual_test_values, predicted_values)
mse = mean_squared_error(actual_test_values, predicted_values)
rmse = np.sqrt(mse)
r2 = r2_score(actual_test_values, predicted_values)


print(f"Mean Absolute Error: {mae:.4f}")
print(f"Mean Squared Error:  {mse:.4f}")
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R^2 Score: {r2:.4f}")

plt.figure(figsize=(12, 4))
plt.plot(actual_test_values, label="Real test data")
plt.plot(predicted_values, label="Predictions")
plt.legend()
plt.title("TCN Prediction vs Real Test Values")
plt.tight_layout()
plt.show()
```
